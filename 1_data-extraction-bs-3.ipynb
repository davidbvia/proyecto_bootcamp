```python
# Almacenar urls de páginas de vehículos

urls_vehiculos = [f"https://www.coches.com/coches-segunda-mano/coches-ocasion-en-barcelona.htm?page={s}" for s in range(0, 501)]
```

```python
%%time

nombres, detalles, contados, financiados = list(), list(), list(), list()

anos, kilometros, combustibles, cambios, potencias, puertas, colores, vendedores = list(), list(), list(), list(), list(), list(), list(), list()

for i in range(0, 501):
    
    url = f"https://www.coches.com/coches-segunda-mano/coches-ocasion-en-barcelona.htm?page={i}"

    response = requests.get(url)

    soup = BeautifulSoup(response.text, "html.parser")

    urls_vehiculos = [s.find("a")["href"] for s in soup.find_all("div", class_ = "Box-module_box__Wgbf3 Card-module_card__Cb1o4 ResultsCardGrid_card__BVUQL CardVO_card__sH3kv Card-module_has-button__R3cTO")]

    for vehiculo in urls_vehiculos:

        response = requests.get(vehiculo)

        soup = BeautifulSoup(response.text, "html.parser")

        ## NOMBRE ########################################
        try:
            nombre = soup.find("div", class_ = "ClassifiedHeader_title__Ym1eO").find("h1").text
        except:
            nombre = np.nan

        ## DETALLES ######################################
        try:
            detalle = soup.find("div", class_ = "ClassifiedHeader_title__Ym1eO").find("span").text
        except:
            detalle = np.nan

        ## PRECIO CONTADO ################################  
        try:
            contado = soup.find("div", class_ = "ClassifiedPriceBlock_classified-price-block__SnJtL").find("strong").text 
        except:
            contado = np.nan

        ## PRECIO FINANCIADO #############################    
        try:
            financiado = soup.find("div", class_ = "ClassifiedPriceBlock_classified-price-block__SnJtL").find("strong").find_next("strong").text
        except:
            financiado = np.nan  

        ## AÑO ###########################################  
        try:
            ano = soup.find("span", class_ = "").text
        except:
            ano = np.nan
        
        ## KILÓMETROS ####################################
        try:
            kilometro = soup.find("span", class_ = "").find_next("span", class_ = "").text
        except:
            kilometro = np.nan
        
        ## COMBUSTIBLES ###################################
        try:
            combustible = soup.find("span", class_ = "").find_next("span", class_ = "").find_next("span", class_ = "").text
        except:
            ano = np.nan

        ## CAMBIOS ########################################
        try:
            cambio = soup.find("span", class_ = "").find_next("span", class_ = "").find_next("span", class_ = "").find_next("span", class_ = "").text
        except:
            cambio = np.nan

        ## POTENCIA #######################################
        try:
            potencia = soup.find("span", class_ = "").find_next("span", class_ = "").find_next("span", class_ = "").find_next("span", class_ = "").find_next("span", class_ = "").text
        except:
            potencia = np.nan
        
        ## PUERTAS ########################################
        try:
            puerta = soup.find("span", class_ = "").find_next("span", class_ = "").find_next("span", class_ = "").find_next("span", class_ = "").find_next("span", class_ = "").find_next("span", class_ = "").text
        except:
            puerta = np.nan

        ## COLOR ##########################################
        try:
            color = soup.find("span", class_ = "").find_next("span", class_ = "").find_next("span", class_ = "").find_next("span", class_ = "").find_next("span", class_ = "").find_next("span", class_ = "").find_next("span", class_ = "").text
        except:
            color = np.nan

        ## TIPO DE VENDEDOR ###############################  
        try:
            vendedor = soup.find("span", class_ = "").find_next("span", class_ = "").find_next("span", class_ = "").find_next("span", class_ = "").find_next("span", class_ = "").find_next("span", class_ = "").find_next("span", class_ = "").find_next("span", class_ = "").text
        except:
            vendedor = np.nan

        
        nombres.append(nombre)
        detalles.append(detalle)
        contados.append(contado)
        financiados.append(financiado)
        anos.append(ano)
        kilometros.append(kilometro)
        combustibles.append(combustible)
        cambios.append(cambio)
        potencias.append(potencia)        
        puertas.append(puerta)
        colores.append(color)
        vendedores.append(vendedor)

        sleep(1)

# Generar dataframe con los datos

df = pd.DataFrame()

df["Nombre"] = nombres 
df["Detalles"] = detalles 
df["Precio al contado"] = contados
df["Precio financiado"] = financiados
df["Año matriculación"] = anos
df["Kilometros"] = kilometros
df["Tipo combustible"] = combustibles 
df["Tipo de cambio"] = cambios 
df["Potencia"] = potencias 
df["Puertas"] = puertas  
df["Tipo de vendedor"] = vendedores 

df.to_csv("coches_segunda_mano.csv")
```
